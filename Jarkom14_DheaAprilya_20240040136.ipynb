```python
import pandas as pd
import string

ciphertext = "YEEIAL BECCS SECCM QICCJYR JRKWK BIICIF ZIJWKQEDAU SECCM WMUWP"
key = "PERCAYALAH"
alphabet = string.ascii_uppercase

plaintext = ""
key_index = 0
table_data = []
no = 1

for char in ciphertext:
    if char == " ":
        plaintext += " "
        key_index = 0   # reset key setiap kata
        continue
    
    c_val = alphabet.index(char)
    k_char = key[key_index % len(key)]
    k_val = alphabet.index(k_char)
    
    p_val = (c_val - k_val) % 26
    p_char = alphabet[p_val]
    
    plaintext += p_char
    
    table_data.append([
        no,
        char,
        k_char,
        c_val,
        k_val,
        f"({c_val} - {k_val}) mod 26",
        p_val,
        p_char
    ])
    
    key_index += 1
    no += 1

table = pd.DataFrame(
    table_data,
    columns=[
        "No",
        "Cipher",
        "Key",
        "Cipher (angka)",
        "Key (angka)",
        "Perhitungan",
        "Hasil (angka)",
        "Plain"
    ]
)

print("Plaintext:")
print(plaintext)

table
```
